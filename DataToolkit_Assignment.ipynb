**1. What is NumPy, and why is it widely used in Python?**

**2. How does broadcasting work in NumPy?**

Broadcasting automatically expands the smaller array during operations to match the dimensions of the larger array.

```python
import numpy as np
a = np.array([1, 2, 3])
b = 2
result = a + b
print('Result of broadcasting:', result)
```

**7. What is a heatmap, and when should it be used?**

A heatmap visualizes matrix-like data with colors representing values. It's used for correlation matrices or to show intensity patterns.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': ['Yes', 'No', 'Maybe']
}
df = pd.DataFrame(data)

# Select only numeric columns
df_numeric = df.select_dtypes(include='number')

# Calculate correlation matrix
correlation_matrix = df_numeric.corr()

# Plot heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Heatmap Example")
plt.show()
```

**18. What are some advanced features of NumPy?**



*   Broadcasting: Allows operations on arrays of different shapes.
*   Universal Functions (ufuncs): Element-wise operations on arrays.
*   Masked Arrays: Handle invalid or missing data.
*   Structured Arrays: Work with complex records.




```python
import numpy as np

# Broadcasting
a = np.array([1, 2, 3])
b = np.array([[4], [5], [6]])
broadcasted_sum = a + b
print("Broadcasted Sum:\n", broadcasted_sum)

# Universal Function
squared = np.square(a)
print("Squared Values:", squared)
```

**19. How does Pandas simplify time series analysis?**

Pandas supports:

*   Datetime indexing for easier slicing and filtering.
*   Resampling for aggregating data (e.g., daily to monthly).
*   Date ranges and frequency generation for time-series data.







```python
import pandas as pd

# Create a time series
date_range = pd.date_range(start="2023-01-01", periods=5, freq="D")
time_series = pd.Series([10, 20, 30, 40, 50], index=date_range)
print("Time Series:\n", time_series)

# Resampling (daily to weekly sum)
weekly = time_series.resample('W').sum()
print("\nWeekly Resampled Data:\n", weekly)
```

**21. Why is NumPy’s array slicing faster than Python’s list slicing?**

NumPy arrays are implemented in C, making them highly optimized for performance. Memory is stored contiguously, and slicing does not create a new copy, unlike Python lists.

```python
# Array slicing in NumPy
array = np.arange(10)
sliced = array[2:8]
print("Original Array:", array)
print("Sliced Array:", sliced)
```
